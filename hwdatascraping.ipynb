```python
print(rows[0].find_all('p')[0].text)
```

```python
col_heads=[head.text.strip() for head in rows[0].find_all('p')]
```

```python
col_heads=[head.text.strip().replace(' ','_').replace('.','') for head in rows[0].find_all('p')]
```

```python
col_heads
```

```python
data_rows=rows[1:]
data_rows[0]
```

```python
value_rows=[[col.text.strip() for col in row.find_all('td')] for row in data_rows]
```

```python
print(value_rows[5])
```

```python
series_list=list(zip(*value_rows))
```

```python
import pandas as pd
from datetime import datetime
```

```python
print(series_list)
```

```python
df=pd.DataFrame(dict(zip(col_heads, series_list)))
```

```python
df
```

```python
type(df["DATES"])
```

```python
df.columns=["DATES","DOW","Text with Explanation"]
```

```python
df["DATE_MONTH"] = df["DATES"].str.slice(start=0, stop=3)
```

```python
df["DATE_DAY"]= df["DATES"].str.slice(start=-2)
```

```python
df["new_date"]=df["DATE_DAY"].str.cat(df["DATE_MONTH"], sep='-')
```

```python
df.new_date[0:40]=df.new_date[0:40]+"-"+"2018"    # I tried to separate month and day parts of the date to convert it to datetime object.
df.new_date[40]="01-Jan-2019"        # Because there are exceptions like intervals "August 27 –September 02", it cannot be directly converted.
```
